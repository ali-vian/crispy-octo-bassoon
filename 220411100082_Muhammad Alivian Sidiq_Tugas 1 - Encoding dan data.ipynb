#### 1. Load Data dari CSV

#### 3. Menghitung ukuran pemusatan data yaitu nilai mean, median dan mode(modus) pada salah satu data numerik

Data numerik yang digunakan adalah bmi

```python
Mean = Dataset['bmi'].mean()
print("Mean dari bmi :",Mean)
```

```python
Median = Dataset['bmi'].median()
print("Median dari bmi :",Median)
```

```python
Modus = Dataset['bmi'].mode()
print("Modus dari bmi :",Modus[0])
```

#### 4. Menghitung ukuran penyebaran data yaitu range(nilai terbesar dan terkecil), variance dan standar deviasi

```python
minimal = Dataset['bmi'].min()
print("Nilai Minimal bmi",minimal)
```

```python
maksimal = Dataset['bmi'].max()
print("Nilai Maksimal bmi",maksimal)
```

```python
variance = Dataset['bmi'].var()
print("Variance bmi",variance)
```

```python
std_dev = Dataset['bmi'].std()
print("Standar Deviasi",std_dev)
```
